```python
import tkinter as tk
from tkinter import messagebox

#creating main window
window = tk.Tk()
window.title("BMI Calculator")
window.geometry("400x350")

#heading
heading = tk.Label(window, text = "BMI Calculator" , font = ("AArial", 18, "bold"))
heading.pack(pady=10)

#weight label and entry
weight_label = tk.Label(window, text="Enter Weight(kg)")
weight_label.pack()

weight_entry = tk.Entry(window)
weight_entry.pack()

#height label and entry
height_entry = tk.Label(window, text = "Enter Height (m)")
height_entry.pack()

height_entry = tk.Entry(window)
height_entry.pack()

#function to calculate bmi
def calculate_bmi():

    try:
        weight = float(weight_entry.get())
        height = float(height_entry.get())

        bmi = weight / (height ** 2)

        result_label.config(text = f"Your BMI is: {round(bmi,2)}")

        #checking category
        if bmi < 18.5:
            category_label.config(text="Category : Underweight")

        elif bmi < 25:
            category_label.config(text="Category : Normal Weight")

        elif bmi < 30:
            category_label.config(text="Category : Overweight")

        else:
            category_label.config(text="Category : Obese")

    except:
        messagebox.showerror("Error", "Please enter valid number")

#calculate button
calculate_button = tk.Button(window, text = "Calculator BMI", command = calculate_bmi)

calculate_button.pack(pady=15)

#result labels
result_label = tk.Label(window, text="", font=("Arial",12))
result_label.pack()

category_label = tk.Label(window, text="", font=("Arial", 12,"bold"))
category_label.pack()

#running window
window.mainloop()
            

        
```
