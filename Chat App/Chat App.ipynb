```python
import tkinter as tk

# creating main window
window = tk.Tk()

window.title("Chat Application")
window.geometry("500x600")

# chat display area
chat_area = tk.Text(
    window,
    height=20,
    width=50,
    font=("Arial", 12)
)

chat_area.pack(pady=20)

# message entry box
message_entry = tk.Entry(
    window,
    width=40,
    font=("Arial", 14)
)

message_entry.pack(pady=10)

# function for sending messages
def send_message():

    # getting user message
    user_message = message_entry.get().lower().strip()

    # displaying user message
    chat_area.insert(tk.END, f"You : {user_message}\n\n")

    # chatbot replies
    if "hello" in user_message:

        bot_reply = "Hello! How are you?"

    elif "how are you" in user_message:

        bot_reply = "I am fine. What about you?"

    elif "i am fine" in user_message:

        bot_reply = "That's great!"

    elif "your name" in user_message:

        bot_reply = "I am a simple chatbot."

    elif "bye" in user_message:

        bot_reply = "Goodbye! Have a nice day."

    else:

        bot_reply = "I don't understand."

    # displaying bot reply
    chat_area.insert(tk.END, f"Bot : {bot_reply}\n\n")

    # clearing entry box
    message_entry.delete(0, tk.END)

# send button
send_button = tk.Button(
    window,
    text="Send",
    font=("Arial", 12, "bold"),
    command=send_message
)

send_button.pack(pady=10)

# running window
window.mainloop()
```
